```python
import pandas
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
radius_diff = np.linspace(0, 1, 100)
```

```python
print(radius_diff)
```

```python
def calculate_fractional_vol_diff(D):
    fraction_vol_diff =  1 - (1-radius_diff)**D
    return fraction_vol_diff
```

```python
dimension_array = np.linspace(1, 30, 5)

plt.xlabel("Radius Difference(ε)")
plt.ylabel("Fractional Volume Difference")
for D in dimension_array:
    plt.plot(radius_diff, calculate_fractional_vol_diff(D), label='D={}'.format(D))

plt.legend()
```
